## Introduction to Machine Learning

### What is Machine Learning?

Machine Learning is the process of extracting knowledge from data automatically, usually with the goal of making predictions on new, unseen data. A classical example is a spam filter, for which the user keeps labeling incoming mails as either spam or not spam. A machine learning algorithm then 'learns' a prediction model from data that distinguishes spam from normal emails, a model which can predict for new emails whether they are spam or not.

Central to machine learning is the concept of **automatic decision making** from data **without the user specifying explicit rules** how this decision shoould be made.

For the case of emails, the user doesn't provide a list of words or characterisitics that make an email spam. Insteam the use provides examples of spam and non-spam emails that are labeled as such.

The second central concept is **generalization**. The goal of a machine learning model is to predict on new, previously unseen data. In a real-world application, we are not interested in marking an already labeled email as spam or not. Instead, we want to make the user's life easier by automatically classifying new incoming mail.

## Multiclass classification

Multiclass classification means a classification task with more than two classes; e.g., classify a set of images of fruits which may be oranges, apples, or pears. 

Multiclass classification makes the assumption that each sample is assigned to one and only one label: a fruit can be either an apple or a pear but not both at the same time.

## Example: Random Forest for Classifying Digits

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.keys()
```

```python
digits.data.shape
```

```python
# set up the figure
fig = plt.figure(figsize=(20, 20))  # figure size in inches
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

# plot the digits: each image is 8x8 pixels
for i in range(100):
    ax = fig.add_subplot(10, 10, i + 1, xticks=[], yticks=[])
    ax.imshow(digits.images[i], cmap=plt.cm.binary, interpolation='nearest')
    
    # label the image with the target value
    ax.text(0, 7, str(digits.target[i]))
```

```python
from sklearn.cross_validation import train_test_split
from sklearn.ensemble import RandomForestClassifier

Xtrain, Xtest, ytrain, ytest = train_test_split(digits.data, digits.target,
                                                random_state=0)
model = RandomForestClassifier(n_estimators=1000)
model.fit(Xtrain, ytrain)
ypred = model.predict(Xtest)
```

```python
from sklearn import metrics
print(metrics.classification_report(ypred, ytest))
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(ytest, ypred)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

## Multilabel classification

```python
from sklearn.multiclass import OneVsRestClassifier
model2 = OneVsRestClassifier(RandomForestClassifier(n_estimators=1000)).fit(Xtrain, ytrain)
```

```python
ypred2 = model2.predict(Xtest)
print(metrics.classification_report(ypred2, ytest))
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(ytest, ypred2)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

```python
model2.estimators_[0].feature_importances_
```

```python
importances = model2.estimators_[0].feature_importances_
std = np.std([tree.feature_importances_ for tree in model2.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(10):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))
```
